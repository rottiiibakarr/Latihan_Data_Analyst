```python
import pandas as pd
import numpy as np
```

```python
# 1. Read Data & Handling Header Spaces (The First Trap):
# A
df = pd.read_csv('Data/data_nilai1.12expert.csv', sep=';')
```

```python
df.head()
```

```python
# B
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
# 2. Removing Duplicates & Useless Columns:
# A 
df.duplicated(subset=['Nama_Nasabah', 'Tipe_Transaksi', 'Waktu_Transaksi']).sum()
df.drop_duplicates(subset=['Nama_Nasabah', 'Tipe_Transaksi', 'Waktu_Transaksi'])
```

```python
# B 
df.drop(columns=['Catatan_Mesin']) # Menghapus kolom Catatan_Mesin
```

```python
# 3. Cleaning Wrong Data (Text Manipulation Chaining):
# A
df['Nama_Nasabah'] = df['Nama_Nasabah'].str.strip().str.title()
```

```python
# B
df['Tipe_Transaksi'] = df['Tipe_Transaksi'].str.upper()
df['Status'] = df['Status'].str.upper()
df.head()
```

```python
# 4. Cleaning Wrong Format (Complex Formatting):
# B
df['Nominal_Rp'] = df['Nominal_Rp'].str.strip()
df['Nominal_Rp'] = df['Nominal_Rp'].str.replace(',', '')
df.head(5)
```

```python
# B
df['Nominal_Rp'] = df['Nominal_Rp'].replace('-', np.nan)
df.head(5)
```

```python
# C
df['Nominal_Rp'] = df['Nominal_Rp'].astype(float)
```
