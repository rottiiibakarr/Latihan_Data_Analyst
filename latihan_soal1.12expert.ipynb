```python
import pandas as pd
import numpy as np
```

```python
# 1. Read Data & Handling Header Spaces (The First Trap):
# A
df = pd.read_csv('Data/data_nilai1.12expert.csv', sep=';')
```

```python
df.head()
```

```python
# B
df.columns = df.columns.str.strip()
```

```python
df.head()
```

```python
# 2. Removing Duplicates & Useless Columns:
# A 
df.duplicated(subset=['Nama_Nasabah', 'Tipe_Transaksi', 'Waktu_Transaksi']).sum()
df = df.drop_duplicates(subset=['Nama_Nasabah', 'Tipe_Transaksi', 'Waktu_Transaksi'])
"""
Bisa menggunakan df.drop_duplicates(subset=['Nama_Nasabah', 'Tipe_Transaksi', 'Waktu_Transaksi'], inplace=True)
dengan tambahan inplace=True
"""
```

```python
# B 
df.drop(columns=['Catatan_Mesin'], inplace=True) # Menghapus kolom Catatan_Mesin
```

```python
# 3. Cleaning Wrong Data (Text Manipulation Chaining):
# A
df['Nama_Nasabah'] = df['Nama_Nasabah'].str.strip().str.title() # Mengubah tulisan menjadi kapital (per-kata)
```

```python
# B
df['Tipe_Transaksi'] = df['Tipe_Transaksi'].str.strip().str.upper() # Mengubah ke dalam tulisan besar semua
df['Status'] = df['Status'].str.strip().str.upper() # Mengubah ke dalam tulisan besar semua
df.head()
```

```python
# 4. Cleaning Wrong Format (Complex Formatting):
# A
df['Nominal_Rp'] = df['Nominal_Rp'].str.strip().replace('-', np.nan) # Mengganti '-' dengan nilai NaN
df.head(5)
```

```python
# B
df['Nominal_Rp'] = df['Nominal_Rp'].str.replace(',', '') # Mengganti ',' dengan string kosong
df.head(5)
```

```python
# C
df['Nominal_Rp'] = df['Nominal_Rp'].astype(float) # Mengubah tipe data dari string ke float
df.head(5)
```

```python
# D
df['Waktu_Transaksi'] = pd.to_datetime(df['Waktu_Transaksi'], format='mixed', dayfirst=True)
"""
Mengubah format Waktu_Transaksi dengan format digunakan untuk mengonversi kolom berisi berbagai format tanggal yang 
tidak konsisten secara otomatis, dengan memprioritaskan hari sebelum bulan
"""
df.head()
```

```python
# 5. Cleaning Wrong Data (Logical Error):
# A
df['Nominal_Rp'] = df['Nominal_Rp'].abs()
df.head(5)
```

```python
# 6. Cleaning Empty Cells (Domain Logic Imputation):
# A. 
df.fillna({'Nominal_Rp': 0}, inplace=True) # Menyisipkan nilai 0 pada kolom Nominal_Rp dengan nilai kolom yang kosong (tidak ada nilai)
df.head(10)
```

```python
# 7. Final Validation & Simple Data Insight (Tugas Analis):
# A
df = df.reset_index(drop=True) # Mengubah ke indeks awal
df.head()
```
