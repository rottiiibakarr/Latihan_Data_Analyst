```python
import pandas as pd
import numpy as np
```

```python
# 1. Read Data:
# A
df = pd.read_csv('Data/data_nilai1.13expert.csv', sep='|')
```

```python
# B
df.columns = df.columns.str.strip() # Menghapus whitespace / spasi pada setiap kolom
```

```python
# 2. Removing Duplicates:
# A 
df.duplicated(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat']).sum() # Menghitung nilai duplikat 
df.drop_duplicates(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat'], inplace=True)

df['ID_Properti'] = df['ID_Properti'].str.strip()
df['Tipe_Properti'] = df['Tipe_Properti'].str.strip()
df['Luas_Tanah'] = df['Luas_Tanah'].str.strip()
df['Sertifikat'] = df['Sertifikat'].str.strip()


df.head()
```

```python
# 3. Cleaning Text:
# A
df['Tipe_Properti'] = df['Tipe_Properti'].str.title() # Mengubah nilai pada kolom Tipe_Properti menjadi tulisan kapital per-kata
```

```python
# B
df['Sertifikat'] = df['Sertifikat'].str.upper() # Mengubah nilai pada kolom Sertifikat menjadi tulisan kapital semua
df.head(10)
```

```python
# 4. Cleaning Luas_Tanah:
# A
df['Luas_Tanah'] = df['Luas_Tanah'].str.replace('m2', '').str.strip() # Mengganti nilai 'm2' dengan string kosong, lalu menghapus whitespace/spasi
df.head()
```

```python
# B
df['Luas_Tanah'] = df['Luas_Tanah'].astype(float) # Mengubah tipe data pada kolom 'Luas_Tanah' menjadi float
df.head()
```

```python
# C
df['Luas_Tanah'] = df['Luas_Tanah'].abs() # Mengubah nilai pada kolom Luas_Tanah menjadi absolut
df.head(10)
```

```python
# D
median_luas_tanah = df['Luas_Tanah'].median() # Mencari nilai median pada kolom Luas_Tanah
df.fillna({'Luas_Tanah': median_luas_tanah}, inplace=True)
df
```

```python
# 5. Cleaning Harga_Jual
# A
df['Harga_Jual'] = df['Harga_Jual'].str.strip().str.replace('Rp', '') # Menghapus whitespace/spasi, lalu mengganti 'Rp' dengan string kosong
df.head()
```

```python
# B
df['Harga_Jual'] = df['Harga_Jual'].str.replace(',', '.') # Mengganti ',' dengan '.'
df.head()
```

```python
# C
def removeNominal(teks) : # Sebuah fungsi untuk menghapus nominal pada teks
    if type(teks) == str : # Mengecek, apakah type data 'teks' sama dengan string
        if 'Juta' in teks : # Jika sama, lalu mengecek kata 'Juta' didalam teks
            angka_bersih = teks.replace('Juta', '').strip() # Membuat variabel untuk mengganti 'Juta' dengan string kosong, lalu menghapus whitespace/spasi
            return float(angka_bersih) * 1000000 # Lalu mengkonversi tipe data dari string ke float, lalu dikali dengan 1 juta
        elif 'Milyar' in teks :
            angka_bersih = teks.replace('Milyar', '').strip() # sama dengan penjelasan sebelumnya, pada 'Juta'
            return float(angka_bersih) * 1000000000
    return teks

df['Harga_Jual'] = df['Harga_Jual'].apply(removeNominal) # Digunakan untuk menerapkan fungsi kustom ke setiap elemen, baris, atau kolom dalam DataFrame
df.head()
```

```python
# 6. Cleaning Tahun_Bangunan :
# A 
df['Tahun_Bangun'] = pd.to_numeric(df['Tahun_Bangun'], errors='coerce')
"""
Fungsi errors = 'coerce' adalah memaksa data yang gagal diubah ke angka menjadi kosong (NaN)

Saat mengonversi kolom menjadi tipe numerik, jika Pandas menemukan teks atau simbol (seperti tulisan " NaN " atau huruf), 
ia biasanya akan langsung memunculkan error dan menghentikan program.

Solusi adalah coerce pada Pandas mengabaikan error tersebut. Data yang bisa diubah akan menjadi angka, sementara data teks yang 
membandel akan secara paksa diganti menjadi NaN (Not a Number atau missing value resmi).
"""

mode_tahun_bangun = df['Tahun_Bangun'].mode()[0] # Menghitung modus/banyaknya data yang muncul pada kolom Tahun_Bangun
df.fillna({'Tahun_Bangun': mode_tahun_bangun}, inplace=True)

df['Tahun_Bangun'] = df['Tahun_Bangun'].astype(int) # Mengubah tipe data dari kolom Tahun_Bangun (string) ke int
```

```python
# 7. Final Validation :
# A
df = df.reset_index(drop=True) # Mereset indeks yang sebelumnya sudah ada yang terhapus, ke indeks yang diurutkan dari awal lagi
df.info()
df
```
