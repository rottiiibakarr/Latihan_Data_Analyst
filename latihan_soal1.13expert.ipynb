```python
import pandas as pd
import numpy as np
```

```python
# 1. Read Data:
# A
df = pd.read_csv('Data/data_nilai1.13expert.csv', sep='|')
```

```python
# B
df.columns = df.columns.str.strip()
```

```python
# 2. Removing Duplicates:
# A 
df.duplicated(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat']).sum()
df.drop_duplicates(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat'], inplace=True)

df['ID_Properti'] = df['ID_Properti'].str.strip()
df['Tipe_Properti'] = df['Tipe_Properti'].str.strip()
df['Luas_Tanah'] = df['Luas_Tanah'].str.strip()
df['Sertifikat'] = df['Sertifikat'].str.strip()


df.head()
```

```python
# 3. Cleaning Text:
# A
df['Tipe_Properti'] = df['Tipe_Properti'].str.title()
```

```python
# B
df['Sertifikat'] = df['Sertifikat'].str.upper()
df.head(10)
```

```python
# 4. Cleaning Luas_Tanah:
# A
df['Luas_Tanah'] = df['Luas_Tanah'].str.replace('m2', '').str.strip()
df.head()
```

```python
# B
df['Luas_Tanah'] = df['Luas_Tanah'].astype(float)
df.head()
```

```python
# C
df['Luas_Tanah'] = df['Luas_Tanah'].abs()
df.head(10)
```

```python
# D
median_luas_tanah = df['Luas_Tanah'].median()
df.fillna({'Luas_Tanah': median_luas_tanah}, inplace=True)
df
```
