```python
import pandas as pd
```

```python
# 1. Read Data:
# A
df = pd.read_csv('Data/data_nilai1.13expert.csv', sep='|')
```

```python
# B
df.columns = df.columns.str.strip()
```

```python
# 2. Removing Duplicates:
# A 
df.duplicated(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat']).sum()
df.drop_duplicates(subset=['ID_Properti', 'Tipe_Properti', 'Luas_Tanah', 'Sertifikat'], inplace=True)

df['ID_Properti'] = df['ID_Properti'].str.strip()
df['Tipe_Properti'] = df['Tipe_Properti'].str.strip()
df['Luas_Tanah'] = df['Luas_Tanah'].str.strip()
df['Sertifikat'] = df['Sertifikat'].str.strip()


df.head()
```
