```python
import pandas as pd
import numpy as np
```

```python
# 1. Read Data:
# A
df = pd.read_csv('Data/data_nilai1.14expert.csv', sep='\t') # sep='\t' yaitu untuk pemisah berupa tab 
```

```python
# B
df.columns = df.columns.str.strip() # Menghapus whitespace/spasi pada seluruh kolom tabel df
```

```python
# 2. Removing Duplicates:
# A
df.drop_duplicates(subset=['Username', 'Teks_Tweet', 'Tanggal_Posting'], inplace=True)
"""
Perintah soal dengan:
"Cek dan hapus baris yang duplikat secara keseluruhan, lalu strip() nama kolom string untuk membuang spasi gaib.",
sudah dilakukan pada tugas 1 bagian B ya!
"""
df # 
```

```python
# 3. Cleaning Username:
# A
df['Username'] = df['Username'].str.replace('@', '')
df.head()
```

```python
# B
df['Username'] = df['Username'].str.lower()
df.head()
```
